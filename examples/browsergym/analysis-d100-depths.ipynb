```python
import os
```

```python
import re
```

```python
tag = "e100"


exp_names = os.listdir("./results")
exp_names

counter = 0
depths = []
for exp_name in exp_names:
  if tag in exp_name:

    exp_path = os.path.join("./results", exp_name)
    task_names = os.listdir(exp_path)

    for task_name in task_names:
      if task_name.startswith("webarena"):
        task_path = os.path.join(exp_path, task_name)
        timetext = os.path.join(task_path, "time.txt")

        content = open(timetext, "r").read()
        matches = re.findall("env", content)
        depths.append(len(matches))
        print(counter, task_name, len(matches))
        counter += 1

    # break
    # print(exp_name)
```

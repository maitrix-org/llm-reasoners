```python
import os
base_path = "./results/best-of-32/webarena.30"
items = os.listdir(base_path)
pickles = [x for x in items if x.endswith(".pkl")]
```

```python
prompt_tokens = 0
cached_prompt_tokens = 0
completion_tokens = 0

import pickle
for file in pickles:
  if file != "result.pkl":
    response = pickle.load(open(os.path.join(base_path, file), "rb"))
    prompt_tokens += response.usage.prompt_tokens
    completion_tokens += response.usage.completion_tokens
    cached_prompt_tokens += response.usage.prompt_tokens_details.cached_tokens
    print(len(response.choices))
    print(prompt_tokens, cached_prompt_tokens, completion_tokens)
print(f"total: {prompt_tokens + completion_tokens} - prompt: {prompt_tokens} : cached {cached_prompt_tokens} - completion {completion_tokens}")

```
